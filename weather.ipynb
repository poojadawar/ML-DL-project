```python
import pandas as pd
import numpy as np

# Sample DataFrames (Replace with actual data)
df = pd.DataFrame({
    "DATE": pd.date_range(start="2023-01-01", periods=10, freq="D"),
    "NEW_YORK_TEMP": np.random.randint(20, 30, size=10),
    "NEW_YORK_HUMIDITY": np.random.randint(40, 60, size=10)
})

df_bbq = pd.DataFrame({
    "DATE": pd.date_range(start="2023-01-01", periods=10, freq="D"),
    "NEW_YORK_BBQ_weather": np.random.choice(["Good", "Bad"], size=10)
})

# Get user input and clean it
city = input("Enter the city name: ").strip().upper()

# Validate user input
if not city:
    print("Error: No city name provided.")
else:
    # Check if 'DATE' column exists
    if "DATE" not in df.columns:
        print("Error: 'DATE' column is missing from the dataset.")
    else:
        # Find relevant columns for the city
        city_columns = [col for col in df.columns if city in col]

        if not city_columns:
            print(f"No data available for {city}.")
        else:
            # Extract relevant city data
            city_data = df.loc[:, ["DATE"] + city_columns]

            # Check if BBQ weather data is available
            bbq_col = f"{city}_BBQ_weather"
            if bbq_col in df_bbq.columns:
                # Merge BBQ data on DATE
                city_data = city_data.merge(df_bbq[["DATE", bbq_col]], on="DATE", how="left").rename(columns={bbq_col: "BBQ"})
            else:
                city_data["BBQ"] = "No Data"

            # Display first 10 rows
            print(city_data.head(10))
```

```python
sns.pairplot(city_data, hue='BBQ', palette='pastel')
plt.title('Pairplot of Weather Variables')
plt.show()
```
